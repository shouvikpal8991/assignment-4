```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 random numbers using np.random.rand()
num_samples = 10000
random_numbers_np = np.random.rand(num_samples)

# Plot the density histogram of the random numbers
plt.hist(random_numbers_np, bins=50, density=True, alpha=0.6, color='b', label='np.random.rand() Histogram')

# Plot the uniform PDF
uniform_pdf = [1] * 100
plt.plot(np.linspace(0, 1, 100), uniform_pdf, 'r--', label='Uniform PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of np.random.rand() Random Numbers vs. Uniform PDF')
plt.legend()
plt.show()
```
