```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Measure the time to generate 10,000 random numbers using np.random.rand()
start_time = time.time()
num_samples = 10000
random_numbers_np = np.random.rand(num_samples)
np_rand_time = time.time() - start_time

# Plot the density histogram of the random numbers
plt.hist(random_numbers_np, bins=50, density=True, alpha=0.6, color='b', label='np.random.rand() Histogram')

# Plot the uniform PDF
uniform_pdf = [1] * 100
plt.plot(np.linspace(0, 1, 100), uniform_pdf, 'r--', label='Uniform PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of np.random.rand() Random Numbers vs. Uniform PDF')
plt.legend()
plt.show()

print(f"Time taken by np.random.rand() to generate 10,000 uniform deviates: {np_rand_time:.6f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Define the LCG parameters
m = 2**31 - 1    # a large prime number
a = 1103515245   # a commonly used multiplier
c = 12345        # a commonly used increment
seed = 42        # initial seed

def lcg(seed, a, c, m, size):
    """Generate a sequence of pseudorandom numbers using LCG."""
    numbers = np.zeros(size)
    numbers[0] = seed
    for i in range(1, size):
        numbers[i] = (a * numbers[i-1] + c) % m
    return numbers / m  # normalize to [0, 1]

# Measure the time to generate 10,000 random numbers using LCG
start_time = time.time()
num_samples = 10000
random_numbers = lcg(seed, a, c, m, num_samples)
lcg_time = time.time() - start_time

# Plot the density histogram of the random numbers
plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g', label='LCG Histogram')

# Plot the uniform PDF
uniform_pdf = [1] * 100
plt.plot(np.linspace(0, 1, 100), uniform_pdf, 'r--', label='Uniform PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of LCG Random Numbers vs. Uniform PDF')
plt.legend()
plt.show()

print(f"Time taken by LCG to generate 10,000 uniform deviates: {lcg_time:.6f} seconds")
```
