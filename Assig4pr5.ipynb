```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Number of samples
num_samples = 5000

# Generate uniform random numbers
u1 = np.random.rand(num_samples)
u2 = np.random.rand(num_samples)
#combining both will generate 10000 random numbers

# Apply Box-Muller transform
z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi* u2)
z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi* u2)

# Combine the two sets of generated numbers
gaussian_numbers = np.concatenate((z0, z1))

# Plot the density histogram of the random numbers
plt.hist(gaussian_numbers, bins=50, density=True, alpha=0.6, color='b', label='Box-Muller Histogram')

# Plot the Gaussian PDF
x = np.linspace(-4, 4, 1000)
gaussian_pdf = norm.pdf(x, 0, 1)
plt.plot(x, gaussian_pdf, 'r--', label='Gaussian PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Box-Muller Random Numbers vs. Gaussian PDF')
plt.legend()
plt.show()
```
