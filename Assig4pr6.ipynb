```python
import numpy as np
import matplotlib.pyplot as plt

# given distribution
def f(x):
    return np.sqrt(2/np.pi) * np.exp(-x**2)

# enveloped function
def g(x):
    if 0 < x < 7:
        return 1
    else:
        return 0

# Define the x range
xarr = np.linspace(0, 10, 10000)

# Calculate f(x) and g(x) values
farr = [f(i) for i in xarr]
garr = [g(i) for i in xarr]

# Rejection sampling
num_samples = 1000
samples_x = []
samples_y = []

while len(samples_x) < num_samples:
    x = np.random.uniform(0, 10)  # Random x between 0 and 10
    y = np.random.uniform(0, 1)   # Random y between 0 and 1
    if y < f(x):
        samples_x.append(x)
        samples_y.append(y)


plt.figure(figsize=(12, 8))
plt.plot(xarr, farr, label='Distribution function f(x)', color='r')
plt.plot(xarr, garr, color='k', label='Enveloped function g(x)')
plt.scatter(samples_x, samples_y, s=1, color='r',alpha=1, label='Random points within envelope')
plt.hist(samples_x, range=(0.0, 10.0), density=True, alpha=0.6, color='g', label='Density')

plt.legend(loc='upper right')
plt.legend()
plt.xlabel('x')
plt.title('Rejection Sampling: Target Distribution and Sampled Data')


plt.show()
```
