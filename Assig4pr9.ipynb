```python
import numpy as np
import matplotlib.pyplot as plt

# Metropolis algorithm to sample from a uniform distribution over (3, 7)
def metropolis_uniform(num_samples, init_x=5, proposal_width=1.0):
    samples = []
    x_current = init_x

    for _ in range(num_samples):
        # Propose a new point
        x_proposal = np.random.normal(x_current, proposal_width)

        # Calculate acceptance probability
        if 3 < x_proposal < 7:
            accept_prob = 1.0
        else:
            accept_prob = 0.0

        # Accept or reject the proposal
        if np.random.rand() < accept_prob:
            x_current = x_proposal

        samples.append(x_current)

    return np.array(samples)

# Parameters
num_samples = 10000
init_x = 5
proposal_width = 0.5

# Generate samples using the Metropolis algorithm
samples = metropolis_uniform(num_samples, init_x, proposal_width)

# Plot the Markov Chain
plt.figure(figsize=(5,3))
plt.plot(samples, lw=0.5)
plt.title('Markov Chain from Metropolis Algorithm')
plt.xlabel('Iteration')
plt.ylabel('Sample value')
plt.show()

# Plot the density histogram
plt.figure(figsize=(5, 3))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot the uniform distribution for comparison
x = np.linspace(3, 7, 1000)
uniform_density = np.ones_like(x) / (7 - 3)
plt.plot(x, uniform_density, 'r-', lw=2, label='Uniform Density')

plt.title('Density Histogram of Samples and Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
